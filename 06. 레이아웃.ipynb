# 06. 레이아웃
- Streamlit에서 레이아웃는 애플리케이션의 외관과 구성을 보다 효율적으로 관리하고 구조화하는데 사용되는 기능
- Streamlit은 기본적으로 단일 페이지를 지원하지만 Streamlit에서 제공하는 기능들을 통해 여러 개의 컬럼으로 레이아웃을 나눌 수 있으며, 각 컬럼에 요소를 배치하여 UI를 효과적으로 구성 가능
- 특정 요소들을 묶어서 스타일링이나 레이아웃 관리 가능

<br>

## 06-01. `st.sidebar`
- 사이드 바
- 작성된 요소들은 왼쪽에 고정
- 사이드바의 크기를 오른쪽 테두리를 드래그 앤 드롭하여 크기를 조정 가능

<br>

### 표기법
- `st.sidebar` 객체 표기법과 `with` 표기법을 사용하여 요소를 전달 가능

<br>

#### `st.sidebar` 객체 표기법

```python
st.sidebar.widget_name
```

<br>

#### `with` 표기법

```python
with st.sidebar:
    st.widget_name
```

<br>

#### 예시
- 사이드 바에 `st.selectbox` 와 `st.checkbox` 위젯과 `st.echo` , `st.spinner` 를 추가한 예시

<br>

```python
import streamlit as st
import time

add_selectbox = st.sidebar.selectbox(
    "어떤 차트를 조회할까요?",
    ("막대", "꺾은선", "히스토그램", "히트맵")
)

with st.sidebar:
    st.checkbox("제목")
    st.checkbox("축제목")
    st.checkbox("눈금선")
    st.checkbox("범례")

with st.sidebar:

    with st.echo():
        st.write("코드 블록 입니다.")
    
    with st.spinner("Loading..."):
        time.sleep(5)

    st.success("끝!")
```

## 06-02. `st.columns`
- `st.columns `는 열(columns) 방향으로 컨테이너를 삽입하여 레이아웃을 만들 수 있는 함수
- 사이드바에서는 칼럼을 삽입할 수 없으며 메인 페이지에만 적용가능

<br>

### 매개변수
- `spec` : 레이아웃을 구성하기 위해 삽입할 열의 수와 너비를 구성하는 매개변수
  - `int` : 정숫값만큼의 개수의 열이 한 행에 만들어 짐
  - `iterable of numbers` : 각각의 열의 너비 비율을 지정
- `gap` : 열 사이의 간격을 조정할 수 있는 매개변수, 지정 값이 없을 경우 기본값은 `small`
  - `small`
  - `medium`
  - `large`

<br>

### 표기법

<br>

#### `st.columns` 객체 표기법

```python
st.columns(spec)[열 인덱스].widget_name
```

<br>

#### `with` 표기법

```python
with st.columns(spec)[열 인덱스]:
  st.widget_name
```

<br>

### 예시
- 2개의 열을 만들어 각 열에 이미지와 텍스트를 배치한 예시


```python
import streamlit as st
from PIL import Image

col1, col2 = st.columns(2, gap="large")

# 컬럼 영역 구분을 위한 css 코드 추가 (파란 가로선)
st.markdown(
  """
  <style>
  .custom-column {
  background-color: lightblue;
  padding: 5px;
  }
  </style>
  """,
  unsafe_allow_html=True
)

with col1:
  st.subheader("column 1")
  image = Image.open("profile.png")
  st.markdown('<div class="custom-column">', unsafe_allow_html=True)
  st.image(image, caption="profile", width=200)

with col2:
  st.subheader("column 2")
  st.markdown('<div class="custom-column">', unsafe_allow_html=True)
  st.header("Jane")
  st.write("Hobby : Soccer")
  st.write("Nice to meet you.")

```

## 06-03. `st.tabs`
- `st.tabs` 는 Streamlit 애플리케이션에 여러 탭 페이지를 생성할 수 있는 기능을 제공
- 한 페이지 내에서 여러 분류를 나누어 구분해서 보여 주는 것이 가능

<br>

### 매개변수
- `tabs` (list of str): 리스트에 포함된 각 문자열마다 탭을 생성
  - 첫 번째 탭은 기본적으로 선택
  - 문자열은 탭의 이름으로 사용되며, 선택적으로 Markdown을 포함 가능
  - Markdown을 사용하여 다음과 같은 요소를 지원: 굵은 텍스트, 이탤릭체(Italics), 취소선, 인라인 코드, 이모지 및 링크

<br>

### 표기법


<br>

#### `st.tabs` 객체 표기법

```python
st.tabs.widget_name
```

<br>

#### `with` 표기법

```python
with st.tabs:
  st.widget_name
```

<br>

### 예시

```python
import streamlit as st
import numpy as np

tab1, tab2 = st.tabs(["선그래프 탭", "데이터 탭"])
data = np.random.randn(10, 3)

with tab1:
  st.subheader("선그래프")
  st.line_chart(data)

with tab2:
  st.subheader("데이터")
  st.dataframe(data)
```

## 06-04. `st.expander`
- 사용자가 펼치거나(확장) 접을(축소) 수 있는 컨테이너
  
<br>

### 매개변수
- `label` (str) : 확장기의 헤더로 사용할 문자열로 접었을 때 표시되는 레이블
- `expanded` (bool) : 기본값은 False(축소)이며, True인 경우 확장 상태로 초기화

<br>

### 표기법

<br>

####  `st.expander` 객체 표기법

```python
expander = st.expander("화살표를 눌러 펼쳐보세요.")
expander.widget_name
```

<br>

#### `with` 표기법

```python
with st.expander:
  st.widget_name
```

<br>

### 예시
- `st.multiselect` 위젯를 추가한 예시

```python
import streamlit as st

colors = ['빨강', '주황', '노랑', '초록', '파랑','보라']

with st.expander("화살표를 눌러 펼쳐보세요."):
  selected_hobbies = st.multiselect('그래프 선 색상을 선택해주세요', colors)
```

## `st.container`
- Streamlit 앱에 여러 요소를 담을 수 있는 가시적이지 않은 컨테이너를 생성하는 함수
- 요소들을 그룹화하고 필요한 시점에 표시하거나 숨길 수 있음

<br>

### 매개변수
- `use_column_width` : 컨테이너의 너비를 현재 열의 너비로 설정할지 여부를 지정
  - 기본값은 `False`
  - `True`로 설정하면 컨테이너의 너비가 현재 열의 너비로 자동 조정
- `scrollable` : 컨테이너가 스크롤 가능한지를 지정
  - 기본값은 `False`
  - `True`로 설정하면 컨테이너가 스크롤 가능
- `expand` : 컨테이너의 너비를 최대로 확장할지 여부를 지정
  - 기본값은 `False`
  - `True`로 설정하면 컨테이너의 너비가 최대로 확장
- `clear_on_reload` : Streamlit 앱이 다시 로드될 때 컨테이너 내부의 상태를 지워야 하는지를 지정
  - 기본값은 `False`
  - `True`로 설정하면 Streamlit 앱이 다시 로드될 때 컨테이너 내부의 상태가 초기화

<br>

### 표기법

<br>

#### `st.container` 객체 표기법

```python
container = st.container("화살표를 눌러 펼쳐보세요.")
container.widget_name
```

<br>

#### `with` 표기법

```python
with st.container:
    st.widget_name
```

<br>

### 예시
- `st.container` 함수를 사용하여 빈 컨테이너를 생성하고, `with` 문을 사용하여 컨테이너 내부에 요소를 추가
- 컨테이너 외부에서도 컨테이너 객체의 메서드를 호출하여 추가 요소

```python
import streamlit as st

# 컨테이너 생성
container = st.container()

# 컨테이너 내부에 다른 컴포넌트들 추가
with container:
    st.header('컨테이너 내부')
    st.subheader('이곳에 다양한 컴포넌트를 추가할 수 있습니다.')
    st.text('여기에는 텍스트, 이미지, 그래프 등 다양한 컴포넌트를 추가할 수 있습니다.')
    st.button('버튼')

# 컨테이너 외부에 다른 컴포넌트들 추가
st.title('컨테이너 외부')
st.subheader('이곳에는 컨테이너 외부의 컴포넌트들이 위치합니다.')
st.text('컨테이너 외부에는 컨테이너 내부의 컴포넌트들과 독립적인 컴포넌트들을 추가할 수 있습니다.')
```

## 06-06. `st.empty`
- 빈 공간을 생성하는 데 사용
- 동적인 내용을 채우기 위해 사용되며, 조건에 따라 다른 컨텐츠를 표시할 때 유용

<br>

### 표기법

<br>

#### `st.empty` 객체 표기법

```python
container = st.empty()
container.위젯_name
```

#### `with` 표기법

```python
with st.empty():
    st.위젯_name
```

<br>

### 예시
- 빈 컨테이너를 생성하고 해당 컨테이너에 텍스트를 삽입하고 시간 간격을 주어 텍스트 내용을 변경하고 삭제

```python
import streamlit as st
import time

st.title("st.empty 예시")

# 빈 컨테이너를 생성합니다.
container = st.empty()

# 컨테이너에 텍스트를 출력합니다.
container.text("Hello, Streamlit!")

# 3초 후에 컨테이너의 내용을 변경합니다.
time.sleep(3)
container.text("Streamlit is awesome🔥")

# 3초 후에 컨테이너의 내용을 clear합니다.
time.sleep(3)
container.empty()
```
